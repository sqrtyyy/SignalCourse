```python
import intelligent_placer
import time
import os
```

```python
total_start = time.time()
result = True
for i, file in enumerate(os.listdir("Data/InArea")):
    start = time.time()
    result = result and intelligent_placer.check_image(f"Data/InArea/{file}")
    print(result)
    print(f"Elapsed: {time.time() - start}")
print("________________________________________")
if result:
    print("All tests passed successfully")
else:
    print("TESTS FAILED")
print(f"Total: {time.time() - total_start}")
print(f"Average: {(time.time() - total_start) / len(os.listdir('Data/InArea'))}")
```

```python
total_start = time.time()
result = False
for i, file in enumerate(os.listdir("Data/OutOfArea")):
    start = time.time()
    result = result or intelligent_placer.check_image(f"Data/OutOfArea/{file}")
    print(result)
    print(f"Elapsed: {time.time() - start}")
print("________________________________________")
if not result:
    print("All tests passed successfully")
else:
    print("TESTS FAILED")
print(f"Total: {time.time() - total_start}")
print(f"Average: {(time.time() - total_start) / len(os.listdir('Data/OutOfArea'))}")
```
